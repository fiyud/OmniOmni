```python
from ultralytics import YOLO, RTDETR
import os
os.environ['KMP_DUPLICATE_LIB_OK']='TRUE'

teacher_model = RTDETR("D:/Eggg/CVR EGG 4.v2i.yolov11/rtdetr-x.pt")

student_model = YOLO("D:/Eggg/CVR EGG 4.v2i.yolov11/yolo-distiller/ultralytics/cfg/models/v8/yolov8s-seg.yaml")

student_model.train(
    data=r"D:/Eggg/CVR EGG 4.v2i.yolov11/data.yaml",
    teacher=teacher_model.model,
    distillation_loss="cwd",
    epochs=150,
    batch=16,
    workers=0,
    exist_ok=True,
)
```

```python
student_model.val(data=r"D:/Eggg/CVR EGG 4.v2i.yolov11/data.yaml", batch=16, device=0, imgsz=640, split="test")
```
